```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
from aruco import *
from label import *
```

```python
path = "games/Bird"
game_num = 0
```

```python
with open(f"{path}.pgn") as pgn:
    for i in range(game_num):
        chess.pgn.skip_game(pgn)
    game = chess.pgn.read_game(pgn)

with open(f"{path}_{game_num}.pkl", "rb") as pkl_wb_obj:
    recorded_moves = pickle.load(pkl_wb_obj)

corners = np.array([corner[0][0] for corner in get_corners(recorded_moves)])
```

```python
img = recorded_moves[0]["color"]
move = list(game.mainline())[0]
```

```python
margin = 0
size = 800

dest = np.float32([[margin,size+margin],[size+margin,size+margin],[margin,margin], [size+margin, margin]])
transform = cv2.getPerspectiveTransform(corners, dest)
board = cv2.warpPerspective(img, transform, (size+2*margin, size+2*margin))
```

```python
def _get_square(square, img, size, margin):
    square_size = size // 8
    i = square % 8
    j = 7 - square // 8
    top_x = i*square_size
    bot_x = (i+1)*square_size + 2*margin
    top_y = j*square_size
    bot_y = (j+1)*square_size + 2*margin
    return img[top_y:bot_y, top_x:bot_x]
```

```python
label = move.board().piece_at(move.move.to_square)
image = _get_square(move.move.to_square, board, size, margin)
move.move.to_square, label
```

```python
plt.imshow(image, interpolation='nearest')
plt.show(), label
```
