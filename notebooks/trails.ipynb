```python
import statistics
all = wesley+rapport+nakamura_f+aranian+quick
stdev = statistics.stdev(all)
mean = statistics.mean(all)

plt.figure(figsize=(10,6))
plt.xlabel('Move Delay (s)')
plt.ylabel('Frequency')
plt.hist(all, bins=50, edgecolor='k')
plt.axvline(mean, color='r', linestyle='dashed')
plt.axvline(mean + stdev, color='y', linestyle='dashed')
plt.savefig('move_delay.png', bbox_inches='tight')
```
