```python
def heights(board, margin=20):
    for square in get_squares(board):
        square = square[margin:-margin,margin:-margin].flatten()
        if height := np.sum(square * (square < 255)):
            yield height
```

```python
sums = np.fromiter(itertools.chain.from_iterable(
    [heights(board(depth))]), np.int64
)
sums
```

```python
jnb = jenkspy.JenksNaturalBreaks(2)
jnb.fit(sums)
jnb.__dict__
```

```python
plt.axhline(jnb.inner_breaks_[0], color="red")
plt.plot(sums)
```

```python
points = sums.tolist()

clusters = []
eps = 200000
points_sorted = sorted(points)
curr_point = points_sorted[0]
curr_cluster = [curr_point]
for point in points_sorted[1:]:
    if point <= curr_point + eps:
        curr_cluster.append(point)
    else:
        clusters.append(curr_cluster)
        curr_cluster = [point]
    curr_point = point
clusters.append(curr_cluster)
len(clusters), clusters
```

```python
kde = KernelDensity(kernel='exponential', bandwidth=2, leaf_size=2).fit(sums.reshape(-1, 1))
e = kde.score_samples(sums.reshape(-1, 1))
```

```python
def occupancy_depth(self, _, depth):
    default = label.LabelOptions()
    board = label.get_board(depth, self.corners)
    depth_colormap = cv2.applyColorMap(cv2.convertScaleAbs(board, alpha=0.03), cv2.COLORMAP_JET)
    square_offset = default.board_size // 16
    for i in range(1, 9):
        for j in range(1, 9):
            x = default.board_size * i // 8 - square_offset
            y = default.board_size * j // 8 - square_offset
            sp = (x - default.cut, y - default.cut)
            ep = (x + default.cut, y + default.cut)
            depth_colormap = cv2.rectangle(depth_colormap, sp, ep, (0, 0, 255), 2)
    self.show_img(depth_colormap, name="depth", size=(500, 500))
    return label.get_occupied_squares(depth, self.corners)
```
